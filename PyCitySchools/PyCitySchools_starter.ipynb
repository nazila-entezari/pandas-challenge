```python
#Calculate the average reading score
average_readingscore=round(school_data_complete['reading_score'].mean(),6)
average_readingscore
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
math_percent=round(len(school_data_complete.loc[school_data_complete['math_score']>=70])/len(school_data_complete)*100,6)
math_percent
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
reading_percent=round(len(school_data_complete.loc[school_data_complete['reading_score']>=70])/len(school_data_complete)*100,6)
reading_percent
```

```python
#Calculate the percentage of students who passed math and reading (% Overall Passing)
passed_df = round(len(school_data_complete.loc[(school_data_complete['reading_score']>=70)&(school_data_complete['math_score']>=70)])/len(school_data_complete)*100,6)
passed_df
```

```python
#creating the table of statistics
stats_df=pd.DataFrame({'Total Schools':[total_schools],
                      "Total Students":[total_students],
                        "Total Budget":[total_budget],
                        "Average Math Score":[average_mathscore],
                       "Average Reading Score":[average_readingscore],
                        "Passing Math":[math_percent],
                       "Passing Reading":[reading_percent],
                       "Overall Passing":[passed_df]
                      })
stats_df
```
