## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
school_summary_df=school_data_complete[["school_name","type","size","budget","math_score","reading_score"]]
```

```python
school_summary_df["Per Student Budget"]=school_summary_df.iloc[:,3]/school_summary_df.iloc[:,2]
```

```python
# % Passing Math
passing_math=[]
previousname=''
school_summary_sorted=school_summary_df.sort_values(["school_name"],ascending=True)
for name in school_summary_sorted.loc[:,"school_name"]: 
    if name!=previousname:
        math=round(len(school_summary_sorted.loc[(school_summary_sorted["school_name"] ==name) & (school_summary_sorted["math_score"]>=70)])/len(school_summary_sorted.loc[(school_summary_df["school_name"] ==name)])*100,5)
        passing_math.append(math)   
    previousname=name    
```

```python
# % Passing readl_summary_sorteding
passing_reading=[]
previousname=''
for name in school_summary_sorted.loc[:,"school_name"]: 
    if name!=previousname:
        reading=round(len(school_summary_sorted.loc[(school_summary_sorted["school_name"] ==name) & (school_summary_sorted["reading_score"]>=70)])/len(school_summary_sorted.loc[(school_summary_df["school_name"] ==name)])*100,5)
        passing_reading.append(reading)  
    previousname=name    
```

```python
# % Passing overall
passing_overall=[]
previousname=''
for name in school_summary_sorted.loc[:,"school_name"]: 
    if name!=previousname:
        overall=round(len(school_summary_sorted.loc[(school_summary_sorted["school_name"] ==name) & (school_summary_sorted["reading_score"]>=70) & (school_summary_sorted["math_score"]>=70)])/len(school_summary_sorted.loc[(school_summary_df["school_name"] ==name)])*100,5)
        passing_overall.append(overall)
    previousname=name    
```

```python
#School Type
previousname=""
stype=[]
y=0
for name in school_summary_sorted.loc[:,"school_name"]:
    while name!=previousname:
        x=len(school_summary_sorted.loc[school_summary_sorted["school_name"]==name,"type"])
        schtype=school_summary_sorted.iloc[y,1] 
        stype.append(schtype)
        y=x+y
        previousname=name 
```

```python
#groupby school
school_summary=school_summary_df.groupby(["school_name"]).mean()
```

```python
#add new columns
school_summary["%Passing Math"]=passing_math
school_summary["%Passing Reading"]=passing_reading
school_summary["%Passing Overall"]=passing_overall
school_summary["School Type"]=stype
```

```python
#Rename the columns
school_summary=school_summary.rename(columns={"size":"Total Students",
                         "budget":"Total School Budget",
                         "math_score":"Average Math Score",
                         "reading_score":"Average Reading Score"})


#rearrange the columns
school_summary=school_summary[["School Type",
                         "Total Students",
                         "Total School Budget",
                         "Per Student Budget",
                         "Average Math Score",
                         "Average Reading Score",
                         "%Passing Math",
                         "%Passing Reading",
                         "%Passing Overall"]]
```

```python
#change the necessary formatting
school_summary_formatted=school_summary.loc[:,:]
school_summary_formatted["Total School Budget"]=school_summary_formatted["Total School Budget"].map("${:,.2f}".format)
school_summary_formatted["Per Student Budget"]=school_summary_formatted["Per Student Budget"].map("${:,.2f}".format)
school_summary_formatted
```

```python
#Group by school
grade_grouped=school_data_complete.groupby(['school_name','grade']).mean().reset_index()
```

```python
#creating the school list
school_name=school_summary.reset_index().iloc[:,0]
# print(school_name)
school_list=[]
for name in school_name:
    school_list.append(name)
```

```python

#Create a pandas series for each grade math

x=0
y=1
z=2
q=3
graded_math_score10=[]
graded_math_score11=[]
graded_math_score12=[]
graded_math_score9=[]
for name in school_list:
    math_by_grade10=grade_grouped.iloc[x:x+1,4]
    graded_math_score10.append(float(math_by_grade10))
    x=x+1
    math_by_grade11=grade_grouped.iloc[y:y+1,4]
    graded_math_score11.append(float(math_by_grade11))
    y=y+1
    math_by_grade12=grade_grouped.iloc[z:z+1,4]
    graded_math_score12.append(float(math_by_grade12))
    z=z+1
    math_by_grade9=grade_grouped.iloc[q:q+1,4]
    graded_math_score9.append(float(math_by_grade9))
    q=q+1

```
